```python
# Import Dependencies
####################################
%matplotlib inline

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
####################################
# Import Market Size Excel File
####################################
FilePath = "Resources/NBA_TeamMarketSize_orig.xlsx"
TeamMarketSize_DF = pd.read_excel(FilePath)
```

```python
# Display Data Frame
####################################
# TeamMarketSize_DF = TeamMarketSize_DF.set_index("Market_Size_Rank")
# print(f'\nTeam Market Size:\n{"*"*30}\n{TeamMarketSize_DF}')
```

```python
# Segment Teams into Three Groups
####################################
Top_Group = TeamMarketSize_DF.loc[1:10 , :]
# print(f'\nTop 10 Teams:\n{"*"*30}\n{Top_Group}')

Middle_Group = TeamMarketSize_DF.loc[11:20 , :]
# print(f'\nMiddle 10 Teams:\n{"*"*30}\n{Middle_Group}')

Bottom_Group = TeamMarketSize_DF.loc[21:30 , :]
# print(f'\nBottom 10 Teams:\n{"*"*30}\n{Bottom_Group}')
```

```python
####################################
# Import Team Valuation Data
####################################
# FilePath = "Resources/TeamValuations_transpose.csv"
FilePath = "Resources/TeamValuations_orig.csv"

TeamValuation_DF = pd.read_csv(FilePath)

# print(f'\nTeam Valuation Data:\n{"*"*30}\n{TeamValuation_DF}')
# print(f'\nTeam Valuation Columns:\n{"*"*30}\n{TeamValuation_DF.columns}')
```

```python
# Clean Data
TeamValuation_DF = TeamValuation_DF.fillna(0)
# print(f'\nTeam Valuation Data:\n{"*"*30}\n{TeamValuation_DF}')

# Rename Columns
TeamValuation_DF.rename(columns = {"Item" : "Teams"} , inplace = True)
# print(f'\nTeam Valuation Col Data Types:\n{"*"*30}\n{TeamValuation_DF.dtypes}')
# print(f'\nTeam Valuation Data:\n{"*"*30}\n{TeamValuation_DF}')
```

```python
TeamValuation_DF.head()
```

```python
Trans_TeamValuation_DF = TeamValuation_DF.T
```

```python
Trans_TeamValuation_DF.head(5)
```

```python
# Transpose Data Frame to make Year Columns and Team Rows

Trans_TeamValuation_DF.columns = Trans_TeamValuation_DF.iloc[0]
Trans_TeamValuation_DF = Trans_TeamValuation_DF.reindex(Trans_TeamValuation_DF.index.drop("Year"))
Trans_TeamValuation_DF.columns = Trans_TeamValuation_DF.columns.values
Trans_TeamValuation_DF.index = [int(x) for x in Trans_TeamValuation_DF.index.values]
```

```python
# Plot the results

Trans_TeamValuation_DF.plot()
plt.grid()

plt.legend(loc = "center left" , bbox_to_anchor = (1.2, .5) , title = "NBA Team")
plt.title("30 NBA Team Valuations" , size = 20)
plt.xlabel("Year", size = 18)
plt.ylabel("Team Valuations (million $)" , size = 15)
plt.gcf().set_size_inches(10 , 8 , forward = True)
```
